```python
#a)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = r"C:\Users\adria\Desktop\Python\Databehandling_AdrianSoderbergSkog\Labs\College.csv"
college = pd.read_csv(path)

college.head()
```

```python
#b)
college2 = pd.read_csv(path, index_col= 0)
college3 = college.rename({"Unnamed: 0": "College"}, axis=1)
college3 = college3.set_index("College")

display(college2.head())

college = college3
display(college.head())
```

```python
#Relationen mellan "grad.rate" och "Outstate", "Expend", "Personal", "perc.alumni", visat i scatterplots

fig, ax = plt.subplots(3, 2, dpi=100, figsize=(10,12))

axes= [(0,0),(0,1),(1,0),(1,1)]
Xcomparison= ["Outstate", "Expend", "Personal", "perc.alumni"]

for ai, xC in zip(axes, Xcomparison):
    sns.scatterplot(data=college, x=xC, y="Grad.Rate", hue= "Elite", size=0.1, alpha=0.6, ax= ax[ai[0], ai[1]])

#Relationen mellan "Outstate" vs "Personal" och "Expeend:"
sns.scatterplot(data=college, x="Outstate", y="Personal", hue= "Elite", size=0.1, alpha=0.6, ax= ax[2, 0])
sns.scatterplot(data=college, x="Outstate", y="Expend", hue= "Elite", size=0.1, alpha=0.6, ax= ax[2, 1])
```
