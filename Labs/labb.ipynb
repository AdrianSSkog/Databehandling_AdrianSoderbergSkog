```python
#a)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = r"C:\Users\adria\Desktop\Python\Databehandling_AdrianSoderbergSkog\Labs\College.csv"
college = pd.read_csv(path)

college.head()
```

```python
#b)
college2 = pd.read_csv(path, index_col= 0)
college3 = college.rename({"Unnamed: 0": "College"}, axis=1)
college3 = college3.set_index("College")

display(college2.head())

college = college3
display(college.head())
```

```python
#c)
college.describe()
```

```python
#d)
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]])
```

```python
#e)
college.boxplot("Outstate", "Private")
```

```python
#f)
college["Elite"] = pd.cut(college["Top10perc"], [0, 50, 100], labels=["no", "yes"])

#college[["Elite", "Top10perc"]].head()
display(college["Elite"].value_counts())

college.boxplot("Outstate", "Elite")
```

```python
#g)
fig, ax = plt.subplots(2,2, dpi= 150, figsize= (10, 8))

columns = ["Outstate", "Room.Board", "Books", "PhD"]
plotIndex = [(0, 0), (0, 1), (1, 0), (1, 1)]
bins = [15, 20, 25, 30]

for c, pI, b in zip(columns, plotIndex, bins):
    college[c].plot.hist(ax=ax[pI[0], pI[1]], bins= b, title= c, alpha= 0.7)
plt.tight_layout()
```

```python
#h)
college[["Personal", "Expend", "Outstate", "Grad.Rate"]].corr()
```

```python
fig, ax = plt.subplots(2,2, dpi= 100, figsize= (8,10))
college.boxplot("Expend", "Elite", ax=ax[0, 0])
college.boxplot("Personal", "Elite", ax=ax[0, 1])
college.boxplot("Grad.Rate", "Elite", ax=ax[1, 0])
college.boxplot("Books", "Elite", ax=ax[1, 1])
```

```python
plt.figure(dpi=100, figsize=(6,4))
sns.scatterplot(data=college, x="Grad.Rate", y="Outstate", hue="Private", size=0.1, alpha=0.6)
```

```python
pd.crosstab(college["Private"], college["Elite"])
```
